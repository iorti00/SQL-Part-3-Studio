### **Write a query that returns tag\_id, the number of times each tag\_id is used and the tag\_name.**  

### Use the GROUP BY and HAVING clause to narrow your focus.

Try multiple keywords, ie woman vs female (female ended up giving better results)

```sql
-- Solution
SELECT TOP 10 COUNT(tag_id) as count_id, goodreads_book_id
FROM dbo.book_tags
GROUP BY goodreads_book_id
```

```sql
SELECT bt.tag_id AS 'TAG ID' ,t.tag_name AS 'TAG NAME',
SUM(bt.count) AS 'TAG COUNT'
FROM book_tags AS bt
INNER JOIN tags AS t 
ON bt.tag_id = t.tag_id
GROUP by bt.tag_id, t.tag_name 
HAVING tag_name LIKE '%female%'
ORDER BY SUM(bt.count) DESC;
```

### **Write a query to return authors, titles, ratings and** **selected tag(s)\_id.**

```sql
-- Solution
SELECT bt.tag_id, b.authors, b.title, b.average_rating
FROM dbo.book_tags AS bt
INNER JOIN dbo.books AS b
ON bt.goodreads_book_id = b.book_id
WHERE bt.tag_id = 11643
ORDER BY b.authors, b.title, b.average_rating DESC
```

```sql
-- Event 1 Query
SELECT bt.tag_id AS 'TAG ID' ,t.tag_name AS 'TAG NAME',
SUM(bt.count) AS 'TAG COUNT'
FROM dbo.book_tags AS bt
INNER JOIN dbo.tags AS t
ON bt.tag_id = t.tag_id
GROUP by bt.tag_id, t.tag_name 
HAVING tag_name LIKE '%cats%' or tag_name LIKE '%cat%'
ORDER BY SUM(bt.count) DESC;
```

```sql
SELECT bk.title AS cat_event, authors, bk.average_rating, bt.tag_id
FROM dbo.books AS bk 
INNER JOIN dbo.book_tags AS bt
ON bk.best_book_id = bt.goodreads_book_id
WHERE  bt.tag_id = 4691 or bt.tag_id = 32186 or bt.tag_id = 6395 or bt.tag_id = 6397
ORDER BY bk.average_rating DESC
```
